```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []

    for entry in entries:
        try:
            username = entry.get('username', '').strip()
            email = entry.get('email', '').strip()
            age = entry.get('age')

            if not username:
                raise ValueError("Empty Username")
            if "@" not in email:
                raise ValueError("Invalid Email")
            if int(age)<=0:
                raise ValueError("Age must be positive")
            valid_entries.append(entry)

        except Exception:
            invalid_entries.append(entry)
    return valid_entries, invalid_entries        
dataset_a = dataset_a
valid_entries, invalid_entries = validate_registrations(dataset_a)

print("Valid entries count:", len(valid_entries))
print("Invalid entries count:", len(invalid_entries))
        
```

Task 3: Row-major vs Column-Major Access Patterns

```python
np.random.seed(42)
n_rows = 200_000
n_cols = 20
columns = [f"feature_{i}" for i in range(n_cols)]
df = pd.DataFrame(
    np.random.rand(n_rows, n_cols),  # values between 0 and 1
    columns=columns
)
print(df.head())
print("\nShape:", df.shape)
```

```python
%time df["feature_0"].mean()
```

```python
%%timeit
total = 0
for i in range(10000):
    total += df.iloc[i]["feature_0"]
```

```python
arr = df.to_numpy()
%time arr[:, 0].mean()
# Expected: ~50-100 microseconds (Fastest)
```

```python
%%timeit
total = 0
for i in range(10000):
    total += arr[i, 0]
# Expected: ~2-5 milliseconds (Much faster than Pandas row access)
```

Why is column access in pandas so much faster than row iteration?
Why does the gap narrow (or change) when using NumPy?
What does this tell you about how pandas and NumPy lay out data in memory?


Pandas is designed for column-wise operations, so accessing an entire column benefits from fast, vectorized computations. Iterating over rows is slower because each row access creates a pandas Series object, introducing additional Python-level overhead.
NumPy arrays are stored in contiguous memory, making both row and column access more efficient compared to pandas. The performance difference arises because NumPy avoids the extra object overhead that pandas incurs.
Overall, both libraries are optimized for column-based operations. NumPy stores data in compact, homogeneous arrays, while pandas builds on NumPy but adds extra structure, which slows down row access.

```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)
rows = 100_000
data = {
    'f1': np.random.randn(rows),
    'f2': np.random.randn(rows),
    'f3': np.random.randn(rows),
    'f4': np.random.randn(rows),
    'f5': np.random.randn(rows),
    'category': np.random.choice(["alpha", "beta", "gamma"], size=rows)
}
df = pd.DataFrame(data)
```

```python
df.to_csv("data.csv", index=False)
df.to_parquet("data.parquet", index=False)

csv_size = os.path.getsize("data.csv") / (1024**2)
pq_size = os.path.getsize("data.parquet") / (1024**2)

print(f"CSV Size: {csv_size:.2f} MB")     # ~11-12 MB
print(f"Parquet Size: {pq_size:.2f} MB") # ~4-5 MB (roughly 60% smaller)
```
