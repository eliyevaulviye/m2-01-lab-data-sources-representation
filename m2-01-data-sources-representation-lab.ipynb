```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []

    for entry in entries:
        try:
            username = entry.get('username',"").strip()
            email = entry.get('email', '').strip()
            age = entry.get('age')

            if not username:
                raise ValueError("Empty username")
            if '@' not in email:
                raise ValueError('Invalid email format')
            if int(age)<=0:
                raise ValueError('Age must be positive')
            
            valid_entries.append(entry)
        except Exception:
            invalid_entries.append(entry)
    return valid_entries, invalid_entries        
dataset_a = dataset_a
valid_entries, invalid_entries = validate_registrations(dataset_a)

print("Valid entries count:", len(valid_entries))
print("Invalid entries count:", len(invalid_entries))
        
```

```python
import pandas as pd 
import numpy as np

np.random.seed(42)
n_rows = 500

rides = pd.DataFrame({
    'ride_id' : np.arange(1, n_rows+1), 
    'driver_id' : np.random.randint(1000, 1051, size=n_rows), 
    'distance_km' : np.random.uniform(1.0, 3.0, size=n_rows),
    'fare_usd': np.random.uniform(5.0, 75.0, size=n_rows), 
    'ride_type': np.random.choice(['standard', 'premium', 'pool']),
    'city': np.random.choice(['Berlin', 'Seoul', 'Nairobi', 'Toronto', 'Lima']),
    'timestamp': pd.date_range(start="2025-01-01", periods=n_rows, freq='T')
})
rides
```
